```python
df = pd.read_csv("ecommerce-data.csv")
df.info()
```

```python
df.info()
```

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
```

```python
df['Month'] = df['OrderDate'].dt.month
df['Year'] = df['OrderDate'].dt.year
```

```python
df.info()
df.head()
```

```python
monthly_sales = df.groupby('Month')['TotalAmount'].sum()
monthly_sales
```

```python
monthly_sales.plot(kind='line', marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

```python
city_sales = df.groupby('City')['TotalAmount'].sum().sort_values(ascending=False)
city_sales
```

```python
city_sales.head(10).plot(kind='bar')
plt.title("Top 10 Cities by Revenue")
plt.xlabel("City")
plt.ylabel("Total Revenue")
plt.show()
```

```python
category_sales = df.groupby('Category')['TotalAmount'].sum().sort_values(ascending=False)
category_sales
```

```python
category_sales.plot(kind='bar')
plt.title("Revenue by Product Category")
plt.xlabel("Category")
plt.ylabel("Total Revenue")
plt.show()
```

```python
top_products = df.groupby('Product')['TotalAmount'].sum().sort_values(ascending=False)
top_products.head(10)
```

```python
top_products.head(10).plot(kind='bar')
plt.title("Top 10 Products by Revenue")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.show()
```

```python
platform_sales = df.groupby('Platform')['TotalAmount'].sum().sort_values(ascending=False)
platform_sales
```

```python
platform_sales.plot(kind='bar')
plt.title("Platform-wise Revenue")
plt.xlabel("Platform")
plt.ylabel("Revenue")
plt.show()
```
